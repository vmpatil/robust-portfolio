```python
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns; sns.set_theme()

from functions import *

import gurobipy as gp
from gurobipy import GRB

import os
import json
import re

import pprint
```

```python
save = True
plot = True

seed = 339

for data in ['GICS','Fama-French']:
    if data == 'GICS':
        unit_list = ['monthly']
        num_sectors_list = [11]
    else:
        unit_list = ['monthly','daily']
        num_sectors_list = [5,10,12,17]
    for units in unit_list:
        for num_sectors in num_sectors_list:

            save_units = units.capitalize()
            
            if units == 'daily':
                N = 100
                num_years=10
                T = 260
                num_units=int(num_years*365*(5/7))
                num_runs = 10
            elif units=='monthly':
                N = 24
                num_years=30
                T = 30
                num_units=int(num_years*12)
                num_runs = 50

            true_mu, true_cov = true_mean_cov(data,units,T,num_sectors,num_units)

            np.random.seed(seed=seed)
            
            r_hat = {}
            mu_hat = {}
            
            for run in range(num_runs):
                r_hat[run] = simulate(true_mu, true_cov, seed)        
                mu_hat[run] = {}
                
                mu_hat[run] = {}
                for t in range(T+N, list(r_hat[run].keys())[-1]-T+1):
                    estimated_expected_return = np.mean(np.array([r_hat[run][i] 
                                                                  for i in range((t-N),t)]), axis=0)
                    mu_hat[run][t] = np.array(estimated_expected_return)

            if data == 'Fama-French':
                path = os.path.join('Simulated Robust','Fama-French',str(num_sectors)+' Sectors',save_units)
            elif data == 'GICS':
                path = os.path.join('Simulated Robust','GICS')
            
            if save:
                incumbent_path = []
                for directory in ['plots'] + os.path.normpath(path).split(os.sep):# + ['Zoomed']:
                    incumbent_path.append(directory)
                    if not os.path.exists(os.path.join(*incumbent_path)):
                        os.mkdir(os.path.join(*incumbent_path))
                incumbent_path = []
                for directory in ['save'] + os.path.normpath(path).split(os.sep):
                    incumbent_path.append(directory)
                    if not os.path.exists(os.path.join(*incumbent_path)):
                        os.mkdir(os.path.join(*incumbent_path))

            min_var_portfolio, min_var_obj = min_variance(true_cov)
            v_begin = min_var_obj
            v_end = np.diag(true_cov).max()
            v_array = np.linspace(v_begin, v_end, 10)
            v_array = [float(np.format_float_positional(v_array[i], precision=3, unique=False, fractional=False, trim='k')) 
                       for i in range(len(v_array))]

            if v_array[0] < min_var_obj:
                v_array[0] = float(str(v_array[0])[:-1] + str(int(str(v_array[0])[-1])+1))
            v_array = v_array[:int(len(v_array)/2)]
            # v_array = [(2/3)*v_begin + (1/3)*v_end]
                        
            equal_portfolio = np.array([1/num_sectors for i in range(num_sectors)])
            equal_weight_variance = equal_portfolio@true_cov@equal_portfolio

            # Xi = Xi_maker(num_sectors, identity=1, sigma=0, inv_sigma=0, power=None, cov=None)
            # Xi_save = 'Identity'
            Xi = Xi_maker(num_sectors, identity=0, sigma=0, inv_sigma=1, power=2, cov=true_cov)
            Xi_save = 'invSigmaSquared'
            # Xi = Xi_maker(num_sectors, identity=0, sigma=0, inv_sigma=1, power=4, cov=true_cov)
            # Xi_save = 'invSigmaFourth'
            # Xi = Xi_maker(num_sectors, identity=0, sigma=0, inv_sigma=1, power=10, cov=true_cov)
            # Xi_save = 'invSigmaTenth'

            print('Xi:')
            display(pd.DataFrame(np.diag(Xi)).T)

            markowitz_portfolios = {}                
            robust_portfolios = {}
            optimal_portfolios = {}
            
            for run in range(num_runs):                
                markowitz_portfolios[str(run)] = {}
                robust_portfolios[str(run)] = {}
                optimal_portfolios[str(run)] = {}
                if run % 10 == 0:
                    print('On run '+str(run))
                    print_idx = True
                else:
                    print_idx = False
                for t in mu_hat[run].keys():
                    
                    if t % 50 == 0 and print_idx:
                        print('\tOn index '+str(t))
                    markowitz_portfolios[str(run)][str(t)] = {}                
                    robust_portfolios[str(run)][str(t)] = {}
                    optimal_portfolios[str(run)][str(t)] = {}
                    
                    # print('Computing Estimated, Robust and Optimal Portfolios')
                    
                    for v in v_array:                      
                        
                        kappa = choose_kappa(mu_hat[run][t],true_cov,Xi,v,lb=1,ub=3)
                        # print('kappa:',kappa)
                        
                        markowitz_portfolio, markowitz_obj = markowitz(mu_hat[run][t], true_cov, v)
                        markowitz_portfolios[str(run)][str(t)][str(v)] = markowitz_portfolio
                        
                        robust_markowitz_portfolio, robust_markowitz_obj = robust(mu_hat[run][t], true_cov, Xi, v, kappa)
                        robust_portfolios[str(run)][str(t)][str(v)] = robust_markowitz_portfolio
                        # print('ratio:',(estimates[j]@robust_markowitz_portfolio) / (kappa*np.sqrt(robust_markowitz_portfolio@Xi@robust_markowitz_portfolio)))
                                                        
                        optimal_portfolio, optimal_obj = markowitz(true_mu[t+T], true_cov, v)
                        optimal_portfolios[str(run)][str(t)][str(v)] = optimal_portfolio
                        
            if save:
                with open(os.path.join('save',path,'markowitz_portfolios (Xi={},N={},{} runs).json'.format(Xi_save,N,num_runs)), 'w') as f: 
                    json.dump(markowitz_portfolios, f, ensure_ascii=False, indent=4)
            
                with open(os.path.join('save',path,'optimal_portfolios (Xi={},N={},{} runs).json'.format(Xi_save,N,num_runs)), 'w') as f: 
                    json.dump(optimal_portfolios, f, ensure_ascii=False, indent=4)
            
                with open(os.path.join('save',path,'robust_portfolios (Xi={},N={},{} runs).json'.format(Xi_save,N,num_runs)), 'w') as f: 
                    json.dump(robust_portfolios, f, ensure_ascii=False, indent=4)

            if save:
                with open(os.path.join('save',path,'markowitz_portfolios (Xi={},N={},{} runs).json'.format(Xi_save,N,num_runs))) as f: 
                    markowitz_portfolios = json.load(f)
            
                with open(os.path.join('save',path,'optimal_portfolios (Xi={},N={},{} runs).json'.format(Xi_save,N,num_runs))) as f: 
                    optimal_portfolios = json.load(f)
            
                with open(os.path.join('save',path,'robust_portfolios (Xi={},N={},{} runs).json'.format(Xi_save,N,num_runs))) as f: 
                    robust_portfolios = json.load(f)
    
            expected_markowitz_returns = {}
            expected_robust_returns = {}
            observed_markowitz_returns = {}
            observed_robust_returns = {}
            optimal_returns = {}
            expected_equal_returns = {}
            observed_equal_returns = {}
            
            for run in range(num_runs):
                expected_markowitz_returns[run] = {}
                expected_robust_returns[run] = {}
                observed_markowitz_returns[run] = {}
                observed_robust_returns[run] = {}
                optimal_returns[run] = {}
                expected_equal_returns[run] = {}
                observed_equal_returns[run] = {}
                
                for t in mu_hat[run].keys():
                    expected_markowitz_returns[run][t] = {}
                    expected_robust_returns[run][t] = {}
                    observed_markowitz_returns[run][t] = {}
                    observed_robust_returns[run][t] = {}
                    optimal_returns[run][t] = {}
                    expected_equal_returns[run][t] = equal_portfolio@mu_hat[run][t]
                    observed_equal_returns[run][t] = equal_portfolio@true_mu[t+T]
    
                    for v in v_array:
                        expected_markowitz_returns[run][t][v] = markowitz_portfolios[str(run)][str(t)][str(v)]@mu_hat[run][t]
                        observed_markowitz_returns[run][t][v] = markowitz_portfolios[str(run)][str(t)][str(v)]@true_mu[t+T]
    
                        expected_robust_returns[run][t][v] = robust_portfolios[str(run)][str(t)][str(v)]@mu_hat[run][t]
                        observed_robust_returns[run][t][v] = robust_portfolios[str(run)][str(t)][str(v)]@true_mu[t+T]
    
                        optimal_returns[run][t][v] = optimal_portfolios[str(run)][str(t)][str(v)]@true_mu[t+T]
    
            mean_expected_markowitz_returns = {}
            mean_expected_robust_returns = {}
            mean_observed_markowitz_returns = {}
            mean_observed_robust_returns = {}
            mean_optimal_returns = {}
            
            for v in v_array:
                mean_expected_markowitz_returns[v] = np.array([expected_markowitz_returns[run][t][v] 
                                                               for t in expected_markowitz_returns[run].keys() 
                                                               for run in range(num_runs)]).mean()
                mean_expected_robust_returns[v] = np.array([expected_robust_returns[run][t][v] 
                                                            for t in expected_robust_returns[run].keys() 
                                                            for run in range(num_runs)]).mean()
                mean_observed_markowitz_returns[v] = np.array([observed_markowitz_returns[run][t][v] 
                                                               for t in observed_markowitz_returns[run].keys()
                                                               for run in range(num_runs)]).mean()
                mean_observed_robust_returns[v] = np.array([observed_robust_returns[run][t][v] 
                                                            for t in observed_robust_returns[run].keys()
                                                            for run in range(num_runs)]).mean()
                mean_optimal_returns[v] = np.array([optimal_returns[run][t][v] 
                                                    for t in optimal_returns[run].keys()
                                                    for run in range(num_runs)]).mean()
                
            mean_expected_equal_returns = np.array([expected_equal_returns[run][t] 
                                                    for t in expected_equal_returns[run] 
                                                    for run in range(num_runs)]).mean()
            mean_observed_equal_returns = np.array([observed_equal_returns[run][t] 
                                                    for t in observed_equal_returns[run] 
                                                    for run in range(num_runs)]).mean()
            if plot:
                font = {'size': 20}
                matplotlib.rc('font', **font)
                matplotlib.rcParams['mathtext.fontset'] = 'stix'
                matplotlib.rcParams['font.family'] = 'STIXGeneral'
                            
                fig, (ax1) = plt.subplots(1,1)
                fig.set_size_inches(21, 12)
                                
                x_title = 'Portfolio Returns'
                color_list = ['black'] + list(sns.color_palette("Set1", n_colors = 2)) + ['slategrey']
                
                ax1.plot(v_array, [mean_optimal_returns[v] for v in v_array], 
                         c=color_list[0], marker='o', linestyle='solid', label='True Frontier')
                
                
                ax1.plot(v_array, [mean_expected_markowitz_returns[v] for v in v_array], c=color_list[1],
                         marker='x', linestyle='solid', label='Estimated Markowitz Frontier')
                ax1.plot(v_array, [mean_observed_markowitz_returns[v] for v in v_array], c=color_list[1],
                         marker='x', linestyle='dashed', label='Actual Markowitz Frontier')
                
                    
                ax1.plot(v_array, [mean_expected_robust_returns[v] for v in v_array], c=color_list[2], 
                         marker='o', linestyle='solid', label='Estimated Robust Frontier')
                ax1.plot(v_array, [mean_observed_robust_returns[v] for v in v_array], c=color_list[2],
                         marker='o', linestyle='dashed', label='Actual Robust Frontier')
                
                equal_weight_v_array = [np.round(equal_weight_variance,1)] + [i for i in v_array if i >= equal_weight_variance]    
                ax1.plot(equal_weight_v_array, [mean_observed_equal_returns for v in range(len(equal_weight_v_array))], 
                         c=color_list[-1], marker='', linestyle='dashed', label='Equal-Weight Frontier')
                
                ax1.set_title(x_title, fontsize=21)
                
                plt.legend(fontsize=18, loc='center left', bbox_to_anchor=(1,0.5))
                
                matplotlib.pyplot.xticks(ticks=v_array)
                
                ax1.set_xlabel('Risk (%)', fontsize=19)
                ax1.set_ylabel('Expected Portfolio Return (%)', fontsize=19)
                
                ax1.tick_params(axis='both', which='major', labelsize=17)
                ax1.tick_params(axis='both', which='minor', labelsize=17)
                
                plt.tight_layout()

                if save:
                    plt.savefig(os.path.join('plots',path,'averaged_robust_frontier_plot (Xi={},N={},{} runs).pdf'.format(Xi_save,N,num_runs)),
                                bbox_inches ='tight')

            gap_closed = {}
                            
            for v in v_array:
                robust_expected_return = mean_observed_robust_returns[v]
                markowitz_expected_return = mean_observed_markowitz_returns[v]        
                true_expected_return = mean_optimal_returns[v]
                gap_closed[v] = (robust_expected_return - markowitz_expected_return)/(true_expected_return - markowitz_expected_return)*100
                
            if not save:
                Xi_save = 'Test'
                
            percent_gap_closed_df = pd.DataFrame(gap_closed, index=['[{}] Gap Closed (%)'.format(Xi_save)])
            display(percent_gap_closed_df)
            
            if save:
                percent_gap_closed_df.to_csv(os.path.join('plots',path,'averaged_gap_closed_table (Xi={},N={},{} runs).csv'.format(Xi_save,N,num_runs)))
                
```
